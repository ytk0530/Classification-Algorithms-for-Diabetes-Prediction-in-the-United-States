# Modelling

```python
from sklearn.model_selection import train_test_split

from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier

from sklearn.metrics import classification_report, confusion_matrix, mean_squared_error
from sklearn import metrics
from sklearn.metrics import accuracy_score

import time
```

```python
def evaluate_model(classifier):
    
    con_mat = confusion_matrix(ytest,classifier.predict(xtest))
    
    
    print("Train Accuracy \n:", accuracy_score(ytrain, classifier.predict(xtrain),'\n'))
    print(f'The Train confusion_matrix is: \n {con_mat}\n')
 
    print("Test Accuracy \n:", accuracy_score(ytest, classifier.predict(xtest),'\n'))
    print(f'The Test confusion_matrix is: \n {con_mat}\n')
    
    print("The classification_report is: \n",classification_report(ytest, classifier.predict(xtest)))
    
    mse = mean_squared_error(ytest,classifier.predict(xtest))
    print('The mean squared error is \n:',mse )
    print('Root mean squared error is \n:', np.sqrt(mse))
```

```python
X= fea_selec.drop('Diabetes_binary',axis=1)
y= df['Diabetes_binary']
```

```python
xtrain, xtest,ytrain, ytest = train_test_split(X, y, train_size =.7, random_state =42)
```

```python
dt = DecisionTreeClassifier()
nb = GaussianNB()
kn = KNeighborsClassifier()
```

```python
dt.fit(xtrain,ytrain)
ypred = dt.predict(xtest)
```

```python
nb.fit(xtrain,ytrain)
ypred = nb.predict(xtest)
```

```python
kn.fit(xtrain,ytrain)
ypred = kn.predict(xtest)
```
